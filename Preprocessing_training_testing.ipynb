```python

PROJECT = Path.cwd()   
DATA_DIR = PROJECT / "data" / "train"


CSV_PATH = DATA_DIR / "_classes.csv"


df = pd.read_csv(CSV_PATH)


print("Rows, Cols:", df.shape)
print(df.head())
print("Columns:", list(df.columns))
```

```python
# Clean column names (remove spaces, slashes, backslashes, lowercase)
orig_cols = list(df.columns)
df.columns = [c.strip().lower().replace('\\', '').replace('/', '').replace(' ', '') for c in df.columns]

print("Before:", orig_cols)
print("After :", list(df.columns))

# Detect the image filename column
image_col_candidates = ['filename', 'image', 'img', 'filepath', 'path']
image_col = next((c for c in image_col_candidates if c in df.columns), None)
assert image_col is not None, f"No image column found. Current columns: {df.columns}"

# Define the class columns
preferred_order = ['freshripe', 'overripe', 'ripe', 'rotten', 'unripe']
class_names = [c for c in preferred_order if c in df.columns]
assert len(class_names) >= 2, f"No valid label columns found. Current columns: {df.columns}"

# Quick check: make sure each row has exactly one label
ones_per_row = df[class_names].sum(axis=1)
print("Rows with exactly one label:", (ones_per_row == 1).mean()*100, "%")
if not (ones_per_row == 1).all():
    print("⚠️ Warning: Some rows do not have exactly one label. Argmax will be used as fallback.")
```

```python
# Step 1: Update class list
preferred_order = ['freshripe', 'freshunripe', 'overripe', 'ripe', 'rotten', 'unripe']
class_names = [c for c in preferred_order if c in df.columns]
print("Classes used:", class_names)

# Step 2: Check how many rows have exactly one label
ones_per_row = df[class_names].sum(axis=1)
print("Rows with exactly one label:", (ones_per_row == 1).mean()*100, "%")

# Step 3: Option A (force argmax): keep all rows but take the max as label
# Option B (clean): drop rows that are not strictly one-hot
# Here is Option B (cleaner dataset):
df_clean = df[ones_per_row == 1].reset_index(drop=True)
print("After cleaning:", df_clean.shape)

# Step 4: Redefine Dataset using the cleaned DataFrame
class BananaDataset(Dataset):
    def __init__(self, dataframe, img_dir, transform=None, image_col='filename', class_names=None):
        self.df = dataframe.reset_index(drop=True)
        self.img_dir = img_dir
        self.transform = transform
        self.image_col = image_col
        self.class_names = class_names

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img_path = self.img_dir / row[self.image_col]
        image = Image.open(img_path).convert("RGB")

        # one-hot -> class index
        label_vec = row[self.class_names].values.astype('float32')
        label_idx = int(label_vec.argmax())

        if self.transform:
            image = self.transform(image)

        return image, label_idx

# Step 5: Transformations and DataLoader
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

train_dir = DATA_DIR
dataset = BananaDataset(df_clean, train_dir, transform=transform, image_col=image_col, class_names=class_names)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# Step 6: Test one batch
images, labels = next(iter(dataloader))
print("Images batch shape:", images.shape)
print("Labels batch shape:", labels.shape)
print("First labels:", labels[:10].tolist())
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self, num_classes=6):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64 * 32 * 32, 256)  # because input is 128x128
        self.fc2 = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)  # flatten
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# create model
model = SimpleCNN(num_classes=len(class_names))
print(model)
```

```python
import torch.nn as nn
import torch.optim as optim

# Loss function (CrossEntropy is best for multi-class classification)
criterion = nn.CrossEntropyLoss()

# Optimizer (Adam works well as a start)
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import matplotlib.pyplot as plt

num_epochs = 10  # try 10 epochs to start
train_loss_history = []
train_acc_history = []

for epoch in range(num_epochs):
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in dataloader:
        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(dataloader)
    epoch_acc = 100 * correct / total

    train_loss_history.append(epoch_loss)
    train_acc_history.append(epoch_acc)

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2f}%")

```

```python
import numpy as np
import torch
from torch.utils.data import DataLoader, WeightedRandomSampler
from sklearn.model_selection import train_test_split

# ---------- 1) Class weights (inverse frequency) ----------
# use float64 and add epsilon to avoid division-by-zero
counts = train_df[class_names].sum().values.astype(np.float64)
eps = 1e-8
class_weights = (counts.sum() / (len(class_names) * (counts + eps)))  # shape: [num_classes]
class_weights_t = torch.tensor(class_weights, dtype=torch.float32)

# ---------- 2) Per-sample weights (vectorized) ----------
labels_idx = train_df[class_names].values.argmax(axis=1)  # shape: [num_samples]

# optional boost for minority class 'unripe'
boost = np.ones(len(class_names), dtype=np.float64)
if 'unripe' in class_names:
    boost[class_names.index('unripe')] = 1.8

sample_weights = class_weights[labels_idx] * boost[labels_idx]          # np.float64
sample_weights = torch.from_numpy(sample_weights).double()              # 1D tensor (CPU)

sampler = WeightedRandomSampler(
    weights=sample_weights,
    num_samples=len(sample_weights),
    replacement=True
)

# ---------- 3) Ensure we have val_df and val_images_dir ----------
if 'val_df' not in globals():
    if 'valid_df' in globals():   # user sometimes names it 'valid_df'
        val_df = valid_df
    else:
        # split from train_df (stratified)
        y_idx = train_df[class_names].values.argmax(axis=1)
        train_df, val_df = train_test_split(
            train_df, test_size=0.2, random_state=42, stratify=y_idx
        )
        # when splitting from train, validation images are still under train_dir
        if 'val_images_dir' not in globals():
            val_images_dir = train_dir

if 'val_images_dir' not in globals():
    val_images_dir = train_dir  # safe fallback

# ---------- 4) Datasets & Dataloaders ----------
train_dataset = BananaDataset(
    train_df, train_dir, transform=train_transform,
    image_col=image_col, class_names=class_names
)

val_dataset = BananaDataset(
    val_df, val_images_dir, transform=val_test_transform,
    image_col=image_col, class_names=class_names
)

# Important: no shuffle when using a sampler
train_loader = DataLoader(train_dataset, batch_size=32, sampler=sampler)
val_loader   = DataLoader(val_dataset,   batch_size=64, shuffle=False)

```

```python
import time, torch
import torch.optim as optim
from tqdm.auto import tqdm

# unfreeze backbone (or keep frozen if you want head-only)
for p in model.parameters():
    p.requires_grad = True

optimizer = optim.Adam(model.parameters(), lr=3e-4)  # small LR for fine-tuning
num_epochs_ft = 5
best_val = float("inf")

print("Starting fine-tuning…")
for epoch in range(1, num_epochs_ft + 1):
    t0 = time.time()

    # ---- Train ----
    model.train()
    train_loss, train_correct, train_total = 0.0, 0, 0
    for imgs, lbls in tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs_ft} [train]"):
        out = model(imgs)
        loss = criterion(out, lbls)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        train_loss += loss.item()
        train_correct += (out.argmax(1) == lbls).sum().item()
        train_total += lbls.size(0)

    # ---- Validate ----
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0
    with torch.no_grad():
        for imgs, lbls in tqdm(val_loader, desc=f"Epoch {epoch}/{num_epochs_ft} [val]"):
            out = model(imgs)
            loss = criterion(out, lbls)
            val_loss += loss.item()
            val_correct += (out.argmax(1) == lbls).sum().item()
            val_total += lbls.size(0)

    tl = train_loss / max(1, len(train_loader))
    ta = 100 * train_correct / max(1, train_total)
    vl = val_loss / max(1, len(val_loader))
    va = 100 * val_correct / max(1, val_total)

    print(f"Epoch {epoch}/{num_epochs_ft} | Train {tl:.4f}/{ta:.2f}% | Val {vl:.4f}/{va:.2f}% | took {time.time()-t0:.1f}s")

    # save best
    if vl < best_val:
        best_val = vl
        torch.save(model.state_dict(), "best_efficientnet_b0.pth")
        print("✓ Saved best_efficientnet_b0.pth")
```
